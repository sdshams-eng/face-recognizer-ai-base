```python
img_gray=cv2.cvtColor(img_cr,cv2.COLOR_BGR2GRAY)
```

```python
plt.close()
plt.imshow(img_gray)
plt.show()
```

```python
cv2.imshow('sa',img_gray)
cv2.waitKey(0)
cv2.destroyWindow()
```

```python
_,img_tr = cv2.threshold(img_gray,90,255,0)
cv2.imshow('sa',img_tr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def ishow(image):
    cv2.imshow('sa',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
ret,img_tr = cv2.threshold(img_gray,90,255,0)
ishow(img_tr)
print(ret)
```

```python
imgc=img2.copy()
cv2.rectangle(imgc,(200,250),(500,480),(0,255,0),3)
cv2.putText(imgc," sajjad shams tonekabon",(100,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
ishow(imgc)
```
